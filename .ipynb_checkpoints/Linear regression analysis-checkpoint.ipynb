```python
sf['Price_per_m2'] = sf['Sold for (SEK)']/sf['Living area (m2)']
corr_matrix = sf.corr()
corr_matrix['Sold for (SEK)'].sort_values(ascending=False)
```

```python
len(sf['Area'].value_counts())
```

```python
freq = sf.groupby('Area').count()['Address']
mean = sf.groupby('Area').mean()['Price_per_m2']
cluster = pd.concat([freq, mean], axis=1)
cluster['Area'] = cluster.index

cluster.columns = ['freq', 'Price_per_m2', 'Area']

cluster.describe()
```

```python
cluster1 = cluster[cluster.Price_per_m2 < 6.078431e+04]
cluster1.index
```

```python
cluster_temp = cluster[cluster.Price_per_m2 >= 6.078431e+04]
cluster2 = cluster_temp[cluster_temp.freq < 123]
cluster2.index
```

```python
cluster3 = cluster_temp[cluster_temp.freq >=123]
cluster3.index
```

```python
def get_group(x):
    if x in cluster1.index:
        return 'low_price'
    elif x in cluster2.index:
        return 'high_price_low_freq'
    else:
        return 'high_price_high_freq'
sf['Group'] = sf.Area.apply(get_group)
```

```python
X = sf[['Longitude', 'Latitude', 'Living area (m2)', 'Rooms', 'Floor', 'Rent', 'LongitudeXLatitude', 'Group']]
Y = sf['Sold for (SEK)']

n = pd.get_dummies(sf.Group)
X = pd.concat([X, n], axis=1)

drops = ['Group']
X.drop(drops, inplace=True, axis=1)
X.head()


X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
print('Linear Regression R squared": %.4f' % regressor.score(X_test, y_test))
```

```python
import numpy as np

from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_pred, y_test)
lin_rmse = np.sqrt(lin_mse)
print('Linear Regression RMSE: %.4f' % lin_rmse)
```

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(y_pred, y_test)
print('Linear Regression MAE: %.4f' % lin_mae)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(X_train, y_train)
```

```python
print('Random Forest R squared": %.4f' % forest_reg.score(X_test, y_test))
```

```python
y_pred = forest_reg.predict(X_test)
forest_mse = mean_squared_error(y_pred, y_test)
forest_rmse = np.sqrt(forest_mse)
print('Random Forest RMSE: %.4f' % forest_rmse)
```

```python
from sklearn import ensemble
from sklearn.ensemble import GradientBoostingRegressor

model = ensemble.GradientBoostingRegressor()
model.fit(X_train, y_train)
```

```python
print('Gradient Boosting R squared": %.4f' % model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
model_mse = mean_squared_error(y_pred, y_test)
model_rmse = np.sqrt(model_mse)
print('Gradient Boosting RMSE: %.4f' % model_rmse)
```

```python
feature_labels = np.array(['Longitude', 'Latitude',
                           'Living area (m2)', 'Rooms', 
                           'Floor', 'Rent', 'LongitudeXLatitude', 'Group'])

importance = model.feature_importances_feature_indexes_by_importance = importance.argsort()

for index in feature_indexes_by_importance:
    print('{}-{:.2f}%'.format(feature_labels[index], (importance[index] *100.0)))
```
